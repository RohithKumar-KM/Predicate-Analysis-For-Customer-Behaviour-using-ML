# Customer Behavior Analysis & Exploratory Data Analysis
## Phase 1


## Objective

The objective of this phase is to explore and understand the customer churn dataset by performing data cleaning, preprocessing, and exploratory data analysis (EDA). The insights obtained will help identify customer behavior patterns and factors influencing churn.

```python
# 1. Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;
```

```python
# 2. Load Dataset
df = pd.read_csv("Customer_churn.csv")
print("="*60)
print("CUSTOMER CHURN DATASET")
print("="*60)
# 3. Dataset Information
print("\nDataset Shape:")
print(df.shape)
print("\nDataset Information:")
df.info()
print("\nFirst Five Records:")
print(df.head());
```

```python
# 3. Missing Value Analysis
print("\n" + "="*60)
print("MISSING VALUES")
print("="*60)
print(df.isnull().sum());
```

```python
# 4.1 Duplicate record analysis
print("Duplicate Records:", df.duplicated().sum());
```

```python
# 4.2 Data Cleaning
df.drop_duplicates(inplace=True)
print("Data cleaned successfully.");
```

```python
# 5. Summary Statistics
print( "="*60)
print("SUMMARY STATISTICS")
print("="*60)
print(df.describe());
```

```python
# 6. Customer Churn Analysis
print("="*60)
print("CHURN DISTRIBUTION")
print("="*60)
print(df["Churn Label"].value_counts());
```

```python
# Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Churn Label", data=df)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show();
```

```python
# Monthly Charges Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Monthly Charges"], bins=30, kde=True)
plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.show();
```

```python
# Gender vs Churn
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", hue="Churn Label", data=df)
plt.title("Gender vs Customer Churn")
plt.show();
```

```python
# Monthly Charges vs Churn
plt.figure(figsize=(8,5))
sns.boxplot(x="Churn Label", y="Monthly Charges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show();
```

```python
# Correlation Heatmap
numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

print("\nBUSINESS INSIGHTS")
print("- Churn distribution shows customer retention patterns.")
print("- Monthly Charges influence churn behaviour.")
print("- Gender-wise churn helps identify demographic trends.")
print("- Correlation analysis identifies important churn factors.");
```

## Conclusion

The exploratory data analysis identified important customer behavior patterns related to churn. These insights provide the foundation for feature engineering and predictive modeling in Phase 2.
